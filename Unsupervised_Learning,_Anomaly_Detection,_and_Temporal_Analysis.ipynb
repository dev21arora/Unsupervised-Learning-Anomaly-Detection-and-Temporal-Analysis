Question 1: What is Dimensionality Reduction? Why is it important in machine learning?

Answer:
Dimensionality reduction is the process of reducing the number of input variables or features in a dataset while retaining the essential information. It transforms high-dimensional data into a lower-dimensional space.

Importance:

Reduces computational cost and training time.

Removes noise and redundant features, improving model performance.

Helps in visualization and better data interpretation.

Prevents the “curse of dimensionality,” which can lead to overfitting.

Question 2: Name and briefly describe three common dimensionality reduction techniques.

Answer:

Principal Component Analysis (PCA): A statistical method that converts correlated variables into a smaller number of uncorrelated variables called principal components, which explain most of the variance.

Linear Discriminant Analysis (LDA): A supervised technique that maximizes class separability while reducing dimensions.

t-Distributed Stochastic Neighbor Embedding (t-SNE): A nonlinear technique mainly used for visualization of high-dimensional data by mapping it to 2D or 3D space.

Question 4: Explain the concept of anomaly detection and its significance.

Answer:
Anomaly detection is the process of identifying data points or events that deviate significantly from the majority of data, indicating potential errors, fraud, or rare events.

Significance:

Helps detect fraud in financial transactions.

Identifies network intrusions and cybersecurity threats.

Monitors equipment for predictive maintenance.

Ensures data quality and system reliability.

Question 5: List and briefly describe three types of anomaly detection techniques.

Answer:

Statistical Methods: Detect anomalies by modeling data distribution and identifying points that fall outside expected probability thresholds.

Machine Learning-Based Methods: Use algorithms such as Isolation Forest or One-Class SVM to learn normal patterns and identify deviations.

Distance-Based Methods: Compute distances between data points; points far from others are considered anomalies (e.g., k-Nearest Neighbors).

Question 6: What is time series analysis? Mention two key components of time series data.

Answer:
Time series analysis involves studying data points collected or recorded at regular time intervals to identify trends, patterns, and seasonal variations.

Key Components:

Trend: The long-term increase or decrease in data over time.

Seasonality: Regular fluctuations that repeat over a specific period (e.g., daily, monthly, yearly).

Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Answer:

Seasonality: Refers to predictable, regular patterns that occur at fixed time intervals (e.g., higher sales during holidays).

Cyclic Behavior: Refers to irregular fluctuations influenced by economic or natural cycles, with varying durations and no fixed period (e.g., business cycles).



Question 8: Write Python code to perform K-means clustering on a sample dataset.


```python
# K-Means Clustering Example
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Create sample data
X, y = make_blobs(n_samples=200, centers=3, cluster_std=0.6, random_state=0)

# Apply K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels = kmeans.labels_

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            color='black', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering Result")
plt.legend()
plt.show()
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.

Answer:
Inheritance in Object-Oriented Programming (OOP) allows one class (child class) to acquire properties and methods of another class (parent class), promoting code reusability and hierarchy.

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Create objects
a = Animal()
d = Dog()

a.speak()  # Output: Animal speaks
d.speak()  # Output: Dog barks
```

Question 10: How can time series analysis be used for anomaly detection?

Answer:
Time series analysis can identify anomalies by detecting sudden deviations from established patterns, trends, or seasonal behavior.

Methods include:

Comparing actual values with forecasted values using models like ARIMA or LSTM.

Detecting spikes or drops in residual errors.

Using statistical thresholds or moving averages to flag unusual data points.

Applications: Fraud detection, stock market monitoring, and equipment failure prediction.
